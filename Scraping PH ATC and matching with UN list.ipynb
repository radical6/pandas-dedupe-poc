```python
import requests
```

```python
import bs4
```

```python
r = requests.get("https://atc.gov.ph/designated-terrorist/")
print(r.content)
```

```python
soup = bs4.BeautifulSoup(r.content,'html.parser')
```

```python
main_content = soup.find('div', attrs = {'class':'entry-content'})
```

```python
t_names = []
for name in main_content.find_all('span', id=False):
    t_names.append(name.text)
```

```python
atc_df = pd.concat([pd.DataFrame({"Name" : pd.Series(t_names).str.strip()}), 
                    pd.concat([orig.T.reset_index().T.set_index(0).T for orig in pd.read_html(str(main_content))], 
                              axis=0).reset_index(drop=True)], 
                   axis = 1)
```

```python
atc_df.to_excel("atc_df.xlsx")
```

```python
atc_df = pd.read_excel("atc_df.xlsx")
```

```python
atc_df['alias_list'] = atc_df['Alias'].replace('Unknown',None).apply(lambda x: x.replace(';','/').split('/') if pd.notna(x) else None)
```

```python
atc_df['variant_list'] = atc_df['Variant'].replace('Unknown',None).apply(lambda x: x.replace(';','/').split('/') if pd.notna(x) else None)
```

```python
atc_pivoted_long = atc_df.rename(columns = {"Unnamed: 0":"ID"}).explode('alias_list').explode('variant_list')
```

```python
uns_individuals["x_CompleteNameLower"] = uns_individuals[['FIRST_NAME','SECOND_NAME','THIRD_NAME','FOURTH_NAME']].apply(lambda x: x.str.strip().str.lower()).agg(lambda y: ' '.join(filter(pd.notna,y)), axis = 1)
uns_individuals["x_TrimmedAliasLower"] = uns_individuals['INDIVIDUAL_ALIAS/0/ALIAS_NAME'].str.strip().str.lower()
```

```python
# uns_entities = pd.read_xml("https://scsanctions.un.org/resources/xml/en/consolidated.xml", xpath = "./ENTITIES/ENTITY")
uns_entities = pd.read_excel("entities_pivoted_long.xlsx")
```

```python
uns_entities["x_CompleteNameLower"] = uns_entities['FIRST_NAME'].str.strip().str.lower()
uns_entities["x_TrimmedAliasLower"] = uns_entities['ENTITY_ALIAS/0/ALIAS_NAME'].str.strip().str.lower()
```

```python
gazetteer_df = pd.concat([uns_individuals[["DATAID","x_CompleteNameLower"]].rename(columns = {"x_CompleteNameLower":"x_Name"}), 
                          uns_individuals[["DATAID","x_TrimmedAliasLower"]].rename(columns = {"x_TrimmedAliasLower":"x_Name"}), 
                          uns_entities[["DATAID","x_CompleteNameLower"]].rename(columns = {"x_CompleteNameLower":"x_Name"}), 
                          uns_entities[["DATAID","x_TrimmedAliasLower"]].rename(columns = {"x_TrimmedAliasLower":"x_Name"})], 
                         axis = 0, ignore_index = True).dropna(axis = 0).drop_duplicates()
```
